```python
# Import necessary libraries
import numpy as np
from sklearn import datasets
from sklearn import neighbors
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Load the iris dataset
iris = datasets.load_iris()

# Print the keys and shape of the iris data
print("Keys of iris dataset: ", iris.keys())
n_samples, n_features = iris.data.shape
print("Shape of data: ", (n_samples, n_features))

# Print the first data point
print("First data point: ", iris.data[0])

# Print the target and target names
print("Target: ", iris.target)
print("Target names: ", iris.target_names)

# Define the indices for the features to be plotted
x_index = 0
y_index = 1

# Create a scatter plot
plt.figure(figsize=(8, 6))
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])
plt.scatter(iris.data[:, x_index], iris.data[:, y_index],
            c=iris.target, cmap=plt.cm.get_cmap('RdYlBu', 3))
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.clim(-0.5, 2.5)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.title("Iris Data Scatter Plot")

# Define the classifier and fit the data
X, y = iris.data, iris.target
clf = neighbors.KNeighborsClassifier(n_neighbors=5)
clf.fit(X, y)

# Predict the class of a new sample
new_sample = [[3, 5, 4, 2]]
result = clf.predict(new_sample)
print("Predicted target name for new sample: ", iris.target_names[result])
```
